# Anatomy of a neural network

* The end goal of a neural networks like most machine learning is to take some input data & produce output data that is desired.
* For eg: Say we want the neural network (NN) to iddentify whether the image is cat or dog. Images of both cat and dog is fed into the NN in pixel form. If its a dog the the final output neuron on top is going to be the strongest, if its a cat then the final neuron at the bottom is going to be strongest
* We can do this by tunning the weights and biases. So all those unique weights and biases, we do that by tunning those and that is the actual training process.
* Tunning those in such a way so that hopefully, we can take data that this NN has never seen, give it pictures of cats & dogs that it has never seen & have it accurately predict those.

* We select one of the neuron and We are going to code a neuron and its somewhere in our neural network
* In a fully connected NN every neuron has a unique connection to every single previous neuron.
* Let say there are 3 neurons that are feeding into this neuron that we are going to build.
* We dont know much about the neuron but we know its going to output some value




```python
# simulating 3 neurons from previous layer feeding into this particular neuron (feed foward)
# we are going to made up some values for our input neurons
inputs = [1.2, 5.1, 2.1]
# Every unique input must have a unique weights associated with it
weights = [3.1, 2.1, 8.7]
# Every unique neuron has a unique bias
bias = 3
# The first step of a neuron is to sum up all the inputs * weights + bias
output = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias
print (output)
```

* The next step we are going to code the layers of the NN
* The weights for now is going to be totally made up
* Later on the NN will randomly initialize weights and then what's going to be tweaking those weights is really like back propagation.
* There is going to be be one bias per neuron that's why we have only 1 neuron
* So for now we are just making the weights up
* The next step we are going to build a neuron shown below




```python
inputs = [1, 2, 3]
# The weights for now is going to be totally made up
weights = [0.2, 0.8, -0.5]
bias = 2
output = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias
print (output)
```

* Now lets add one more input 
* What are inputs, inputs in theory could be one of two things it could be truly an input as in value from the input layer of a neural network which is just vector of values or list of values. Just like our example above, the inputs are from the input layer which feeds into the hidden layer.
* The input layer is simply whatever values you're tracking from various sources such as sensors.
* Later we are going to add another input to we're going to simulate its coming from the middle of the neural network.
* The inputs could be from another neuron or from various sensors.
* In the below example, we are going to take inputs from one of the hidden layer not from sensors as shown in the screenshot below





```python
# let say we add another input
inputs = [1, 2, 3, 2.5]
weights = [0.2, 0.8, -0.5, 1.0]
bias = 2.0

output = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + inputs[3]*weights[3] + bias
print (output)
```

* What if we want to model 3 neurons with 4 inputs



* We are going to model the output layer.
* What we're going to need here if its for inputs for 3 neurons this means the inputs is going to stay the same, we have already have 4 inputs defined.
* What we're going to need or since there's three neurons that means there's going to be three unique weight sets and each weight sets is going to have 4 values because there's 4 inputs and then for biases we are going to need 3 unique and seperate biases.


```python
inputs = [1, 2, 3, 2.5]

weights1 = [0.2, 0.8, -0.5, 1.0]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2.0
bias2 = 3.0
bias3 = 0.5

# make it into a list to simulate output layer
output = [inputs[0]*weights1[0] + inputs[1]*weights1[1] + inputs[2]*weights1[2] + inputs[3]*weights1[3] + bias1,
          inputs[0]*weights2[0] + inputs[1]*weights2[1] + inputs[2]*weights2[2] + inputs[3]*weights2[3] + bias2,
          inputs[0]*weights3[0] + inputs[1]*weights3[1] + inputs[2]*weights3[2] + inputs[3]*weights3[3] + bias3]

print(output)

```
